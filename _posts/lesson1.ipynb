```python
import numpy as np
import matplotlib.pyplot as plt

x_data=[1.0,2,3]
y_data=[2.0,4,6]

def forward(x):#向前函数
    return x*w

def loss(x,y):#计算损失函数
    y_pred=forward(x)
    return (y_pred-y)*(y_pred-y)

w_list=[]
mse_list=[]

for w in np.arange(0.0,4.1,0.1):#生成从0.0到4.1间隔为0.1的数列
    print('w=',w)
    loss_sum=0
    for x_val,y_val in zip(x_data,y_data):#解包x_data和y_data配对迭代器
        y_pred_val=forward(x_val)
        loss_val=loss(x_val,y_val)
        loss_sum+=loss_val#求和
        print('\t',x_val,y_val,y_pred_val,loss_val)
    w_list.append(w)
    mse_list.append(loss_sum/3)
    print('mse=',loss_sum/3)
plt.plot(w_list,mse_list)
plt.ylabel('LOSS')
plt.xlabel('w')
plt.show()
```

```python
import numpy as np
#绘制三维图
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

x_data=[1.0,1.4,3.5]
y_data=[2.0,4,6]

def forward(x):#向前函数
    return x*w+b

def loss(x,y):#计算损失函数
    y_pred=forward(x)
    return (y_pred-y)*(y_pred-y)

w_list=[]
mse_list=[]
b_list=[]

for w in np.arange(0.0,4.1,0.1):#生成从0.0到4.1间隔为0.1的数列
    for b in np.arange(0.0,2.0,0.1):
        print('w=',w)
        print('b=',b)
        loss_sum=0
        for x_val,y_val in zip(x_data,y_data):#解包x_data和y_data配对迭代器
            y_pred_val=forward(x_val)
            loss_val=loss(x_val,y_val)
            loss_sum+=loss_val#求和
            print('\t',x_val,y_val,y_pred_val,loss_val)
        w_list.append(w)
        b_list.append(b)
        mse_list.append(loss_sum/3)
        print('mse=',loss_sum/3)
        
#转化为numpy数组
x = np.array(w_list)
y = np.array(b_list)
z = np.array(mse_list)

#数据网格化
xi = np.linspace(x.min(), x.max(), 100)
yi = np.linspace(y.min(), y.max(), 100)
xi, yi = np.meshgrid(xi, yi)#meshgid函数生成规则网格
zi = griddata((x, y), z, (xi, yi), method='cubic')

# 创建图形和3D坐标轴
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# 绘制曲面图
surf = ax.plot_surface(xi, yi, zi, cmap='viridis', alpha=0.9, 
                       linewidth=0, antialiased=True)

# 添加颜色条
fig.colorbar(surf, shrink=0.5, aspect=5)

# 设置坐标轴标签
ax.set_xlabel('w')
ax.set_ylabel('b')
ax.set_zlabel('mse')

# 显示图形
plt.show()
```
